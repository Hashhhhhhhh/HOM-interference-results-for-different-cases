```python
#Plotting dip depth vs overlap
overlap = np.linspace(0, 1, 100)  # 0 = distinguishable, 1 = indistinguishable
x2 = 0.5 # Transmission probability (50:50 BS)

# Classical and quantum coincidence probabilities
p11_classic = x2**2 + (1 - x2)**2
p11_quantum = p11_classic - 2 * x2 * (1 - x2) * overlap  # General formula

# HOM dip depth (visibility)
dip_depth = p11_quantum / p11_classic

plt.plot(overlap, 1 - dip_depth, label=f'Transmission x={x2}')
plt.xlabel('Photon Overlap')
plt.ylabel('Dip Depth (Visibility)')
plt.title('HOM Dip Depth vs. Photon Overlap')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
overlap = np.linspace(0, 1, 100)  # 0 = distinguishable, 1 = indistinguishable
x2 = 0.5  # Transmission probability (50:50 BS)

# Classical and quantum coincidence probabilities
p11_classic = x2**2 + (1 - x2)**2
p11_quantum = p11_classic - 2 * x2 * (1 - x2)*overlap

# HOM dip depth (visibility)
dip_depth = p11_quantum / p11_classic

plt.plot(overlap, dip_depth, label=f'Transmission x={x2}')
plt.xlabel('Photon Overlap')
plt.ylabel('Normalized Coincidence Probability')
plt.title('HOM Dip Depth vs. Photon Overlap')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#normalized coincidental probability vs delays
# Parameters
tau = np.linspace(-2, 2, 200)  # Time delay (in units of coherence time)
sigma = 0.5  # Wavepacket width (coherence time)
x2 = 0.5     # 50:50 beam splitter transmission coefficient

# gaussian wavepackets overlap of as a function of delay
overlap = np.exp(-tau**2 / (2 * sigma**2))

# Normalized Coincidence probability
p11_classic = x2**2 + (1 - x2)**2
p11_quantum = p11_classic - 2 * x2 * (1 - x2) * overlap

plt.plot(tau, p11_quantum / p11_classic, label='HOM Dip (Gaussian)')
plt.xlabel('Delay (units of coherence time)')
plt.ylabel('Normalized Coincidence Probability')
plt.title('HOM Dip for Gaussian Wavepackets')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#normalized coincidence prob vs beam splitter transmission prob

x2_values = np.linspace(0, 1, 100)
overlap = 0.5  # Perfect indistinguishability

p11_classic = x2_values**2 + (1 - x2_values)**2
p11_quantum = p11_classic - 2 * x2_values * (1 - x2_values) * overlap

plt.plot(x2_values, p11_quantum / p11_classic, label='HOM Dip Depth')
plt.xlabel('Beam Splitter Transmission Probability (x)')
plt.ylabel('Normalized Coincidence Probability')
plt.title('HOM Dip Depth vs. Beam Splitter Ratio')
plt.grid(True)
plt.legend()
plt.show()
```
